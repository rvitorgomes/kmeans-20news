```python
import pickle
# Need the interactive Tools for Matplotlib
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
 
from sklearn.manifold import TSNE
```

```python
# LOAD THE DATASET
data_f = open('./representations/tokens_vector_abstracts.pkl', 'rb')
voc = pickle.load(data_f)
print(len(voc), 'tokens loaded')

data_f = open('./representations/documents_abstracts.pkl', 'rb')
docs = pickle.load(data_f)
print(len(docs), 'tokens loaded')

print('reading...')
X = np.loadtxt('./representations/tfidf_matrix_abstracts.txt')
print('loaded')
print(X.shape)
```

```python
# find tsne coords for 2 dimensions
tsne = TSNE(n_components=2)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(X)
```

```python
# Display the word2vec
plt.figure()

x_coords = Y[:, 0]
y_coords = Y[:, 1]
# display scatter plot
plt.scatter(x_coords, y_coords)

for label, x, y in zip(voc, x_coords, y_coords):
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
    
plt.rcParams['figure.figsize'] = (9, 9)
        
plt.xlim(x_coords.min()+1.05, x_coords.max()+1.05)
plt.ylim(y_coords.min()+1.05, y_coords.max()+1.05)
plt.title('Word Embeddings Visualization')
plt.show()
```
